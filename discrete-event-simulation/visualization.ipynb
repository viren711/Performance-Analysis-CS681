```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('../discrete-event-simulation/users_results_UNI_FIFO.csv')

# Create the plot with proper labels and grid
# plt.figure(figsize=(12, 7))
plt.plot(df['users'], df['throughput'], 'o-', linewidth=3, markersize=10, color='darkblue', label='Throughput')

# Add labels and title
plt.xlabel('Number of Users', fontsize=14)
plt.ylabel('Throughput (req/s)', fontsize=14)
plt.title('System Throughput vs Users', fontsize=16, fontweight='bold', pad=20)

# Customize grid
plt.grid(True, alpha=0.4, linestyle='--', linewidth=1)
plt.minorticks_on()
# plt.grid(which='major', alpha=0.4, linestyle='--', linewidth=1)
# plt.grid(which='minor', alpha=0.2, linestyle=':', linewidth=0.8)

# Set proper ticks
plt.xticks(df['users'])
# plt.yticks(range(0, int(df['throughput'].max()) + 1, 1))

# Add legend
plt.legend(loc='best', fontsize=12)

# Improve layout
plt.tight_layout()
# plt.savefig('labeled_throughput.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# plt.figure(figsize=(10, 6))
plt.plot(df['users'], df['utilization']*100, 'o-', linewidth=2.5, markersize=8, color='red')
plt.xlabel('Number of Users', fontsize=12,  )
plt.ylabel('CPU Utilization (%)', fontsize=12)
plt.title('CPU Utilization vs Users', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(df['users'])
plt.tight_layout()
plt.savefig('cpu_utilization.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
# plt.savefig('labeled_throughput.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# plt.figure(figsize=(10, 6))
plt.plot(df['users'], df['mean_response'], 's-', linewidth=2.5, markersize=8, 
         color='green')
plt.xlabel('Number of Users', fontsize=12)
plt.ylabel('Response Time (seconds)', fontsize=12)
plt.title('Response Time vs Users', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(df['users'])
plt.legend()
plt.tight_layout()
plt.show
```

```python
plt.plot(df['users'], df['drop'], 'D-', linewidth=2.5, markersize=8, 
         color='orange', label='Request Drop Fraction')
plt.xlabel('Number of Users', fontsize=12)
plt.ylabel('Drop Fraction', fontsize=12)
plt.title('Request Drop Rate vs Users', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(df['users'])
plt.ylim(0, 0.01) # Zoom to show zero line clearly
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(df['throughput'], df['utilization']*100, 'o-', linewidth=2.5, markersize=8, 
         color='purple', label='CPU Utilization')
plt.xlabel('Throughput', fontsize=12)
plt.ylabel('CPU Utilization (%)', fontsize=12)
plt.title('CPU Utilization vs Throughput', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.ylim(0, 100)  # Y-axis starts at 0
plt.legend()
plt.tight_layout()
# plt.savefig('cpu_vs_throughput.png', dpi=300, bbox_inches='tight')
plt.show()
```

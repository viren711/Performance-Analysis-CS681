```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("user_results_EXP_FIFO.csv")

plt.plot(data["users"], data["mean_response"])
plt.xlabel("Users")
plt.ylabel("Response-Time")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sim = pd.read_csv("users_results_EXP_FIFO.csv")
mva = pd.read_csv("mva_EXP_FIFO.csv")

df = pd.merge(sim, mva, on="users")

plt.figure(figsize=(8,5))

plt.plot(df["users"], df["mean_response"],
         marker='o', label="Simulation")

plt.plot(df["users"], df["R_mva"],
         marker='x', label="MVA")

plt.xlabel("Number of Users")
plt.ylabel("Response Time")
plt.title("Response Time: Simulation vs MVA")
plt.legend()
plt.grid()

plt.savefig("response_comparison.png")
plt.show()


plt.figure(figsize=(8,5))

plt.plot(df["users"], df["throughput"],
         marker='o', label="Simulation")

plt.plot(df["users"], df["X_mva"],
         marker='x', label="MVA")

plt.xlabel("Number of Users")
plt.ylabel("Throughput")
plt.title("Throughput: Simulation vs MVA")
plt.legend()
plt.grid()

plt.savefig("throughput_comparison.png")
plt.show()
```
